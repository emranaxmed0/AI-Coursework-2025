```python
days_str = input ("Enter number of days: ")
try:
    days = int(days_str)
    if days < 0:
        print("Please enter a non-negative number of days.")
    else:
        seconds = days * 24 * 60 * 60
        print(f"{days} day(s) = {seconds} seconds")
except ValueError:
    print("Invalid input. Please enter a whole number.")
```

```python
import math

r_str = input("Enter radius of the sphere: ")

try:
    r = float(r_str)
    if r < 0:
        print("Radius cannot be naegative.")
    else:
        volume = (4/3) * math.pi * (r ** 3)
        print (f"Volume = {volume}")
except ValueError:
    print("Invalid input. Please enter a numeric radius.")
```

```python
def square_area(side: float) -> float:
    return side * side

def square_perimeter(side: float) -> float:
    return 4 *side

side_str = input("Enter side length of the square: ")

try:
    side = float(side_str)
    if side < 0:
        print("Side lenght cannot be negative.")
    else:
        print(f"Area = {square_area(side)}")
        print(f"perimeter = {square_perimeter(side)}")
except ValueError:
    print("Invalid input. Please enter a numeric value.")
    
```

```python
def case_of_char(ch: str) -> str:
    if len(ch) != 1:
        return "Please enter exactly one character."
    if ch.isupper():
        return "UPPERCASE"
    elif ch.islower():
        return "lowercase"
    else:
        return "Neither uppercase nor lowercase (e.g., a digit or symbol)."

char = input("Enter a single character: ")
print(case_of_char(char))
```

```python
x = 0
y = 20

while True:
    y -= 4
    x += 2 / y
    if y < 6:
        break

print(f"x = {x}")
```

```python
values = []
count_needed = 5

print(f"Enter {count_needed} numeric values: ")
while len(values) < count_needed:
    s = input(f"value {len(values)+1}: ")
    try:
        values.append(float(s))
    except ValueError:
        print("Not a number-try again.")

average = sum(values) / len(values)
print(f"Values: {values}")
print(f"Average = {average}")
                            
```
